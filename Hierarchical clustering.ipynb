```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
X = np.array([[1, 3],
             [2, 4],
             [1, 5],
             [5, 5],
             [5, 7],
             [4, 9],
             [2, 8],
             [3, 10],
            ])
n, m = X.shape 
```

```python
from math import dist

D = np.zeros([n, n])
for i in range(n):
    for j in range(n):
        D[i, j] = dist(X[i], X[j])
np.round(D, 3)
```

Lo anterior es la forma de calcurar la matriz $D$, sin embargo utilizaremos las librer√≠as para hacer el dendograma y el cluster analysis.

```python
dendrogram = sch.dendrogram(sch.linkage(X, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
y_hc = hc.fit_predict(X)
plt.scatter(X[y_hc==0, 0], X[y_hc == 0, 1], s=50, color='red')
plt.scatter(X[y_hc==1, 0], X[y_hc == 1, 1], s=50, color='blue')
plt.scatter(X[y_hc==2, 0], X[y_hc == 2, 1], s=50, color='green')
# plt.scatter(X[y_hc==3, 0], X[y_hc == 3, 1], s=50, color='yellow')
```

```python
X = np.random.random([50, 2])
```

$\text{single-linkage}$

```python
dendrogram = sch.dendrogram(sch.linkage(X, method='single'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='single')
y_hc = hc.fit_predict(X)
plt.scatter(X[y_hc==0, 0], X[y_hc == 0, 1], s=50, color='red')
plt.scatter(X[y_hc==1, 0], X[y_hc == 1, 1], s=50, color='blue')
plt.scatter(X[y_hc==2, 0], X[y_hc == 2, 1], s=50, color='green')
```

```python
dendrogram = sch.dendrogram(sch.linkage(X, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')
y_hc = hc.fit_predict(X)
plt.scatter(X[y_hc==0, 0], X[y_hc == 0, 1], s=50, color='red')
plt.scatter(X[y_hc==1, 0], X[y_hc == 1, 1], s=50, color='blue')
plt.scatter(X[y_hc==2, 0], X[y_hc == 2, 1], s=50, color='green')
plt.scatter(X[y_hc==3, 0], X[y_hc == 3, 1], s=50, color='yellow')
plt.scatter(X[y_hc==4, 0], X[y_hc == 4, 1], s=50, color='pink')
```
